```
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```
# defining function to avoid repetition

def rang(obj,name):
    df = obj[name].describe()
    print("The range for " + name + " is between %.2f and %.2f" % (df[3], df[7]))

def mstd(obj, name):
    df = obj[name].describe()
    print("The mean and standard deviation for " + name + " is %.2f and %.2f, respectively" % (df[1], df[2]))

def printData(auto):
  print("\n<--------------- Range of predictors ------------------->")
  rang(auto, 'mpg')
  rang(auto, 'cylinders')
  rang(auto, 'displacement')
  rang(auto, 'horsepower')
  rang(auto, 'weight')
  rang(auto, 'acceleration')
  rang(auto, 'year')
  print("\n<--------------- Mean and Standard deviation of predictors ------------------->")
  mstd(auto, 'mpg')
  mstd(auto, 'cylinders')
  mstd(auto, 'displacement')
  mstd(auto, 'horsepower')
  mstd(auto, 'weight')
  mstd(auto, 'acceleration')
  mstd(auto, 'year')
```

```
# import csv
auto = pd.read_csv('Auto.csv')
print("\n<--------------- Displaying information ------------------->")
print(auto)

# removing column name
auto = auto.drop(['name', 'origin'], axis=1)

# create backup to later create scatter plot of all the data
bk_auto = auto
```

```
# print range, mean and standard deviation of data
printData(auto)
```

```
# make a list of indx observations you want to remove
dlist = [i for i in range(9,84)]

# remove observations from data
auto = auto.drop(dlist)
print("\n<--------------- Displaying removed information ------------------->")
print(auto)

# print range, mean and standar deviation after removing from 10th - 85th observation
printData(auto)
```

```
sb.pairplot(bk_auto)
plt.show()
```
